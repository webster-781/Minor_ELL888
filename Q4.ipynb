```python
R,I,C,y,df,ind = getAdultData(5000)
```

```python
RM,IM,CM,yM = getAdultData2(amount = 1000, col = 'Sex', set= ' Male')
RF,IF,CF,yF = getAdultData2(amount = 1000, col = 'Sex', set= ' Female')
DM = getAllDist(RM,IM,CM)
DF = getAllDist(RF,IF,CF)
effM = np.array(all_effs(DM,yM))
effF = np.array(all_effs(DF,yF))
print(effM.shape)
eff = (effM + effF)/2
```

```python
print(eff.shape)
```

```python
K = list(range(1,11))
plt.plot(K,eff[0],color = 'green')
plt.plot(K,eff[1], color = 'blue')

plt.xlabel("Number of Neighbours(k)")
plt.ylabel("Efficiency of Graph")
plt.title("Constructing different graphs for different Genders(Male and Female), then assembling")
plt.legend(['Neighbourhood Efficiency','Classification Efficiency'])
plt.show()
```
